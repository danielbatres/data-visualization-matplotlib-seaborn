```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 30)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(9, 5))

axes[0].plot(x, y, label='$y = sin(x)$')
axes[0].set_title('$y = cos(x)$')
axes[0].set_xlabel('$x$')
axes[0].set_ylabel('$y$')
axes[0].legend()

axes[1].plot(y, x, label='$x = sin(y)$')
axes[1].set_title('$x = cos(y)$')
axes[1].set_xlabel('$y$')
axes[1].set_ylabel('$x$')
axes[1].legend()
```

```python
plt.plot(x, y, label='$y = sin(x)$')
plt.title('This is a title')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left', bbox_to_anchor=(0.5, 0.2))
```
