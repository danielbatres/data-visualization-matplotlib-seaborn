```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1, 50, 100)
```

## Histogram

```python
plt.hist(data, bins=10, histtype='bar')
plt.show()
```

```python
plt.hist(data, bins=10, histtype='step')
plt.show()
```

## Boxplot

```python
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True)
plt.show()
```

## Scatter

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n)) ** 2
colors = np.random.rand(n)
```

```python
plt.scatter(x, y, s=area, c=colors, marker='p')
plt.show()
```
