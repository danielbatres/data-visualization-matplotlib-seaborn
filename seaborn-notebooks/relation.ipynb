```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.scatterplot(tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
plt.figure(figsize=(8, 8))
markers = { 'Lunch': 'D', 'Dinner': 's' }
sns.scatterplot(tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
sns.lineplot(tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.relplot(tips, x='total_bill', y='tip', hue='day', style='time', size='size', col='time')
plt.show()
```
