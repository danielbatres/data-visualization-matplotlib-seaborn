```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5, 5, 25)
y = x ** 2
```

```python
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
plt.plot(x, y, 'rs-')
plt.show()
```

```python
plt.plot(x, y, 'yD:')
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(y)
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```
